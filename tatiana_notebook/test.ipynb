## Linear Regression with Scatter plot and line plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from pathlib import Path

def load_and_preprocess(file_path):
    df = pd.read_csv(file_path)
    df = df.dropna()
    df['date'] = pd.to_datetime(df['date'])
    df['date_numeric'] = df['date'].view('int64')
    return df

file_paths = [
    Path('../data/store1_data.csv'),
    Path('../data/store2_data.csv'),
    Path('../data/store3_data.csv'),
    Path('../data/store4_data.csv'),
    Path('../data/store5_data.csv')
]

store_data = [load_and_preprocess(fp) for fp in file_paths]

colors = ['blue', 'green', 'red', 'purple', 'orange']

plt.figure(figsize=(14, 8))

for i, (df, color) in enumerate(zip(store_data, colors), start=1):
    x = df['date_numeric'].values.reshape(-1, 1)
    y = df['Sales'].values.reshape(-1, 1)

    model = LinearRegression()
    model.fit(x, y)
    y_pred = model.predict(x)

    plt.scatter(df['date'], df['Sales'], label='Store ' + str(i) + ' Data Points', color=color, alpha=0.6)
    plt.plot(df['date'], y_pred, label='Store ' + str(i) + ' Regression Line', color=color, linewidth=2)

plt.xlabel('Date')
plt.ylabel('Sales in Gallons')
plt.title('Sales Forecasting - Linear Regression for Multiple Stores')
plt.legend()
plt.show()
```
